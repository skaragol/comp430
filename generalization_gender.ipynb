```python
import pandas as pd

df = pd.read_csv("lungcancer.csv", sep=";")

with open("gender.txt", "r") as file:
    gender_mapping = dict(line.strip().split(';') for line in file)

df['GENDER'] = df['GENDER'].map(gender_mapping)

df.to_csv("lungcancer_gender_generalized.csv", sep=";", index=False)

df.head(20)
```
