```python
import pandas as pd
import numpy as np

df = pd.read_csv("lungcancer.csv", sep=";")

def perturb_age(age):
    noise = np.random.uniform(low=-2, high=2)
    perturbed_age = int(round(int(age) + noise))
    return perturbed_age

df['AGE'] = df['AGE'].apply(perturb_age)

df.to_csv("lungcancer_perturbed.csv", sep=";", index=False)

df.head(20)
```
